```python
import numpy as np
import pandas as pd
import scipy.special
import scipy.stats as st
from math import sqrt
from scipy.stats import beta

from matplotlib import pyplot as plt
plt.style.use('seaborn-white') 
```

```python
plt.figure(figsize=(6, 4), dpi=160)

x = np.linspace(0, 1, 400, endpoint=True)
beta_prior = st.beta.pdf(x, 8, 5) 
beta_post = st.beta.pdf(x, 17, 8)

plt.xlim(0.0, 1.0)
plt.ylim(0.0, 5.0)

plt.plot(x, beta_post, color=dark_red, linewidth=1.5, linestyle="-", label="$\pi(\\theta|x)$")
plt.plot(x, beta_prior, color=dark_blue, linewidth=1.5, linestyle="-", label="$\pi(\\theta)$")

plt.xticks([0, 1], ['$0$', '$1$'])
plt.yticks([])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.legend(loc='upper right', fontsize=10)
plt.savefig("beta_pres_post.png", dpi=160)
```
